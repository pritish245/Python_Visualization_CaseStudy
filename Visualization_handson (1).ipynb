```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# import dataset
data=pd.read_csv("cars-1 (2) (1).csv")
data.head()
```

```python
data.info()
```

```python
# convert these columns into numeric
data['cubicinches']=pd.to_numeric(data['cubicinches'],errors='coerce')
data['weightlbs']=pd.to_numeric(data['weightlbs'],errors='coerce')
data.info()
```

Create scatter plots for the following data, make sure all the plots appear in the same
plane with x-labels and y-labels evenly spaced.
- Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg,
hp-time-to-60) with respect to each brand.


```python
fig,axes=plt.subplots(2,2,figsize=(20,10))

plt.subplot(2,2,1)
sns.scatterplot(x='mpg',y='hp',hue='brand',data=data)

plt.subplot(2,2,2)
sns.scatterplot(x=data['hp'],y=data['time-to-60'],hue=data['brand'],data=data)

plt.subplot(2,2,3)
sns.scatterplot(x=data['hp'],y=data['mpg'],hue=data['brand'],data=data)


plt.subplot(2,2,4)
sns.scatterplot(x=data['mpg'],y=data['weightlbs'],hue=data['brand'],data=data)
```
